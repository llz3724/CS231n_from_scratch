## Answer:
The five-layer network was noticeably more sensitive to the initialization scale compared to the three-layer network.

Observations:

Training Difficulty: The five-layer network often failed to converge or learned very slowly when the initialization scale was not optimal, whereas the three-layer network was more robust to suboptimal initialization scales.

Sensitivity to Initialization: Small changes in the initialization scale (e.g., using He initialization vs. Xavier initialization, or slightly scaling the weights up/down) had a much larger impact on the five-layer network's performance. The three-layer network, in contrast, was more forgiving.

Why?

Vanishing/Exploding Gradients: In deeper networks (like the five-layer one), gradients can vanish or explode more easily during backpropagation if the initialization scale is not properly tuned. This is because the gradients are multiplied through more layers, amplifying small initialization errors.

Greater Depth, Greater Sensitivity: Each additional layer introduces more parameters and nonlinearities, making the optimization landscape more complex. Poor initialization can lead to activations that are too small (causing vanishing gradients) or too large (causing exploding gradients), both of which hinder learning.

Three-Layer Robustness: With fewer layers, the gradients are less likely to vanish or explode, and the network can often recover from suboptimal initialization through training.

## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer:
Why AdaGrad Updates Become Very Small

AdaGrad's update rule accumulates the square of all past gradients (cache += dw) and uses this to adapt the learning rate for each parameter. The key issue is that the cache grows monotonically over time, causing the effective learning rate to decay:

Cache Accumulation:

As training progresses, cache keeps increasing because it sums all historical squared gradients (cache += $dw^2$).

The denominator in the update rule (np.sqrt(cache) + eps) grows larger, making the effective step size (learning_rate / np.sqrt(cache)) smaller and smaller.

Resulting Behavior:

Early in training, the updates are large because cache is small.

Later in training, cache dominates, and the updates become extremely small, leading to slow convergence or premature stopping.

Problem Cases:

In deep networks or long training runs, AdaGrad can become too conservative, effectively stopping learning prematurely.

This is especially problematic for sparse features, where rare but important updates might get "drowned out" by the accumulated cache.

Would Adam Have the Same Issue?

No, Adam does not suffer from the same problem because it uses exponentially moving averages (EMA) of past gradients (m_t for momentum) and squared gradients (v_t for scaling), rather than a monotonically increasing sum like AdaGrad.

The update is scaled by m_t / ($\sqrt{v_t}$ + eps), where v_t is an EMA rather than a sum.

Why Adam Avoids the Problem:

Forgetting Old Gradients: Since v_t is an EMA (controlled by ), it discards old gradient information exponentially. This prevents the denominator from growing indefinitely.

Adaptive but Non-Decaying Learning Rate: The effective learning rate stabilizes over time rather than decaying to near zero.

Better for Long Training: Adam remains effective even in long training runs, unlike AdaGrad.

Trade-offs:

Adam can sometimes converge too aggressively early in training (due to bias correction), but it generally avoids AdaGrad's vanishing-updates issue.
